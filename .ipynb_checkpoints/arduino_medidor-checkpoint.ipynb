```python
import serial
import pandas as pd
from datetime import datetime, timedelta
from IPython.display import display, clear_output, HTML
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import PolynomialFeatures, MinMaxScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.covariance import EllipticEnvelope
from statsmodels.tsa.seasonal import STL
from fpdf import FPDF
import joblib
import warnings
warnings.filterwarnings('ignore')
```

# InicializaciÃ³n (conexiÃ³n y estructuras)

```python
puerto = 'COM3'
baudrate = 115200
arduino = serial.Serial(puerto, baudrate, timeout=1)

# ParÃ¡metros ajustables
config = {
    'modelos': {
        'lineal': make_pipeline(MinMaxScaler(), LinearRegression()),
        'polinomial': make_pipeline(PolynomialFeatures(degree=3), MinMaxScaler(), LinearRegression()),
        'random_forest': RandomForestRegressor(n_estimators=100, max_depth=4, random_state=42),
        'gradient_boost': GradientBoostingRegressor(n_estimators=50, random_state=42)
    },
    'modelo_actual': 'gradient_boost',
    'intervalo_prediccion': 60,  # 60 segundos de predicciÃ³n
    'paso_prediccion': 5,  # cada 5 segundos
    'ventana_analisis': 30,  # Ãºltimos 30 puntos para anÃ¡lisis
    'umbral_anomalias': 0.05,
    'update_frecuency': 1  # actualizar cada 1 lectura
}

# Variables globales
datos = []
ultima_actualizacion = datetime.now()
estadisticas = {
    'consumo_promedio': 0,
    'max_consumo': 0,
    'min_consumo': 0,
    'tendencia': 'estable'
}

# ConexiÃ³n SQL Server en Python
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=LAPTOP-IOFJQ6TD\SQLEXPRESS'
)
cursor = conn.cursor()

def insertar_evento_sql(fila):
    cursor.execute("""
        INSERT INTO EnergiaEventos 
        (Timestamp, TiempoEventoSegundos, EnergiaEventoWh, 
         TiempoTotalSegundos, EnergiaTotalWh, EsAnomalia)
        VALUES (?, ?, ?, ?, ?, ?)
    """, (
        datetime.strptime(fila['Timestamp'], '%Y-%m-%d %H:%M:%S.%f'),
        fila['Tiempo Evento (s)'],
        fila['Energia Evento (Wh)'],
        fila['Tiempo Total (s)'],
        fila['Energia Total (Wh)'],
        int(fila['Anomalia'])
    ))
    conn.commit()
```

# ConversiÃ³n de tiempo a segundos

```python
def tiempo_a_segundos(t):
    """Convierte tiempo HH:MM:SS a segundos con manejo de errores"""
    try:
        h, m, s = map(int, t.split(":"))
        return h * 3600 + m * 60 + s
    except:
        return 0

def analizar_tendencia(serie):
    """Determina la tendencia actual usando regresiÃ³n lineal"""
    if len(serie) < 3:
        return "insuficiente datos"
    
    X = np.arange(len(serie)).reshape(-1, 1)
    lr = LinearRegression().fit(X, serie)
    pendiente = lr.coef_[0]
    
    if pendiente > 0.1:
        return "â†‘ ascendente"
    elif pendiente < -0.1:
        return "â†“ descendente"
    else:
        return "â†’ estable"

def detectar_anomalias(df):
    """DetecciÃ³n avanzada de anomalÃ­as con mÃºltiples mÃ©todos"""
    if len(df) < 10:
        df['Anomalia'] = False
        return df
    
    # MÃ©todo 1: Envelope elÃ­ptico
    X = df[['Tiempo Evento (s)', 'Energia Evento (Wh)']].values
    detector = EllipticEnvelope(contamination=config['umbral_anomalias'])
    df['Anomalia_1'] = detector.fit_predict(X) == -1
    
    # MÃ©todo 2: DesviaciÃ³n del promedio mÃ³vil
    window_size = max(3, len(df)//5)
    df['Media_Movil'] = df['Energia Evento (Wh)'].rolling(window=window_size, center=True).mean()
    df['Std_Movil'] = df['Energia Evento (Wh)'].rolling(window=window_size, center=True).std()
    df['Anomalia_2'] = abs(df['Energia Evento (Wh)'] - df['Media_Movil']) > 2 * df['Std_Movil']
    
    # Combinar resultados
    df['Anomalia'] = df['Anomalia_1'] | df['Anomalia_2']
    return df.drop(columns=['Anomalia_1', 'Anomalia_2', 'Media_Movil', 'Std_Movil'])

def calcular_estadisticas(df):
    """Calcula estadÃ­sticas avanzadas del consumo"""
    if len(df) == 0:
        return estadisticas
    
    ventana = df.tail(config['ventana_analisis'])
    estadisticas['consumo_promedio'] = ventana['Energia Evento (Wh)'].mean()
    estadisticas['max_consumo'] = ventana['Energia Evento (Wh)'].max()
    estadisticas['min_consumo'] = ventana['Energia Evento (Wh)'].min()
    estadisticas['tendencia'] = analizar_tendencia(ventana['Energia Evento (Wh)'].values)
    
    return estadisticas

def generar_predicciones(modelo, df):
    """Genera predicciones detalladas a corto plazo"""
    if len(df) < 5:
        return None, None, {}
    
    X = np.array(df['Tiempo Evento (s)']).reshape(-1, 1)
    y = df['Energia Evento (Wh)']
    
    # Entrenar modelo
    modelo.fit(X, y)
    
    # Generar predicciones
    ultimo_tiempo = df['Tiempo Evento (s)'].iloc[-1]
    tiempos_pred = np.arange(
        ultimo_tiempo, 
        ultimo_tiempo + config['intervalo_prediccion'], 
        config['paso_prediccion']
    ).reshape(-1, 1)
    
    pred = modelo.predict(tiempos_pred)
    
    # Calcular mÃ©tricas
    pred_train = modelo.predict(X)
    mae = mean_absolute_error(y, pred_train)
    rmse = np.sqrt(mean_squared_error(y, pred_train))
    r2 = r2_score(y, pred_train)
    
    metricas = {
        'MAE': f"{mae:.4f} Wh",
        'RMSE': f"{rmse:.4f} Wh",
        'RÂ²': f"{r2:.4f}",
        'Tendencia': estadisticas['tendencia']
    }
    
    return tiempos_pred, pred, metricas

def crear_tabla_estilo(df):
    """Crea una tabla HTML con estilo para mejor visualizaciÃ³n"""
    styled_df = df.tail(10).style \
        .background_gradient(subset=['Energia Evento (Wh)'], cmap='YlOrRd') \
        .bar(subset=['Energia Total (Wh)'], color='#5fba7d') \
        .apply(lambda x: ['background: #ffcccc' if v else '' for v in x], subset=['Anomalia'])
    
    return styled_df
```

# Lectura, anÃ¡lisis y predicciÃ³n en tiempo real

```python
try:
    contador = 0
    while True:
        linea = arduino.readline().decode('utf-8').strip()
        
        if linea.startswith("REPORTE"):
            partes = linea.split(", ")
            if len(partes) == 5:
                # Procesamiento avanzado de datos
                timestamp = datetime.now()
                fila = {
                    'Timestamp': timestamp.strftime('%Y-%m-%d %H:%M:%S.%f')[:-3],
                    'Tiempo Evento': partes[1],
                    'Energia Evento (Wh)': float(partes[2]),
                    'Tiempo Total': partes[3],
                    'Energia Total (Wh)': float(partes[4]),
                    'Delta_Tiempo': 0
                }
                
                if len(datos) > 0:
                    ultimo = datos[-1]['Timestamp']
                    delta = (timestamp - datetime.strptime(ultimo, '%Y-%m-%d %H:%M:%S.%f')).total_seconds()
                    fila['Delta_Tiempo'] = delta
                
                datos.append(fila)
                df = pd.DataFrame(datos)
                
                # ConversiÃ³n y anÃ¡lisis avanzado
                df['Tiempo Evento (s)'] = df['Tiempo Evento'].apply(tiempo_a_segundos)
                df['Tiempo Total (s)'] = df['Tiempo Total'].apply(tiempo_a_segundos)
                df = detectar_anomalias(df)
                calcular_estadisticas(df)

                insertar_evento_sql(fila)  # Insertar en SQL Server
                
                # Guardado periÃ³dico
                if contador % 10 == 0:
                    df.to_csv("data_dash.csv", index=False)  # Para el dashboard
                    df.to_csv("datos_energia_luz.csv", index=False)
                    joblib.dump(config['modelos'][config['modelo_actual']], 'modelo_actual.pkl')
                
                # ActualizaciÃ³n visual mejorada
                if contador % config['update_frecuency'] == 0:
                    clear_output(wait=True)
                    
                    # Generar predicciones
                    tiempos_pred, energias_pred, metricas = generar_predicciones(
                        config['modelos'][config['modelo_actual']],
                        df
                    )
                    
                    # Mostrar informaciÃ³n detallada
                    print("â•"*70)
                    print(f"âš¡ SISTEMA DE MONITOREO ENERGÃ‰TICO EN TIEMPO REAL âš¡".center(70))
                    print("â•"*70)
                    
                    # InformaciÃ³n de estado
                    print(f"\nðŸ“Š ESTADO ACTUAL (Ãºltimos {min(config['ventana_analisis'], len(df))} puntos):")
                    print(f"  â€¢ Consumo promedio: {estadisticas['consumo_promedio']:.3f} Wh")
                    print(f"  â€¢ MÃ¡ximo consumo: {estadisticas['max_consumo']:.3f} Wh")
                    print(f"  â€¢ MÃ­nimo consumo: {estadisticas['min_consumo']:.3f} Wh")
                    print(f"  â€¢ Tendencia: {estadisticas['tendencia']}")
                    print(f"  â€¢ AnomalÃ­as detectadas: {sum(df['Anomalia'])}/{len(df)}")
                    
                    # Mostrar tabla con estilo
                    display(crear_tabla_estilo(df))
                    
                    # Mostrar mÃ©tricas del modelo
                    if metricas:
                        print("\nðŸ“ˆ MÃ‰TRICAS DEL MODELO:")
                        for k, v in metricas.items():
                            print(f"  â€¢ {k}: {v}")
                    
                    # Mostrar predicciones
                    if tiempos_pred is not None:
                        print(f"\nðŸ”® PREDICCIÃ“N PARA PRÃ“XIMOS {config['intervalo_prediccion']} SEGUNDOS:")
                        for i, (t, e) in enumerate(zip(tiempos_pred.flatten(), energias_pred)):
                            t_rel = t - df['Tiempo Evento (s)'].iloc[-1]
                            print(f"  â€¢ +{t_rel:02d}s: {e:.3f} Wh", end=" | ")
                            if (i+1) % 3 == 0:
                                print()
                    
                    # VisualizaciÃ³n avanzada
                    plt.figure(figsize=(15, 10))
                    
                    # GrÃ¡fico 1: Serie temporal con predicciÃ³n
                    plt.subplot(2, 2, 1)
                    ventana = df.tail(config['ventana_analisis'])
                    
                    plt.plot(ventana['Tiempo Evento (s)'], ventana['Energia Evento (Wh)'], 
                            'b-o', label='Consumo real', markersize=5)
                    
                    if tiempos_pred is not None:
                        plt.plot(tiempos_pred, energias_pred, 'r--', label='PredicciÃ³n')
                        plt.scatter(tiempos_pred, energias_pred, c='red', s=30, marker='x')
                    
                    anomalias = ventana[ventana['Anomalia']]
                    if not anomalias.empty:
                        plt.scatter(anomalias['Tiempo Evento (s)'], anomalias['Energia Evento (Wh)'], 
                                   c='black', s=100, marker='s', label='AnomalÃ­as')
                    
                    plt.title(f"Consumo y PredicciÃ³n (Modelo: {config['modelo_actual'].upper()})")
                    plt.xlabel("Tiempo (segundos)")
                    plt.ylabel("EnergÃ­a (Wh)")
                    plt.legend()
                    plt.grid(True, linestyle='--', alpha=0.7)
                    
                    # GrÃ¡fico 2: Histograma de consumo
                    plt.subplot(2, 2, 2)
                    plt.hist(df['Energia Evento (Wh)'], bins=15, color='skyblue', edgecolor='black')
                    plt.axvline(estadisticas['consumo_promedio'], color='red', linestyle='--', 
                               label=f'Promedio: {estadisticas["consumo_promedio"]:.2f} Wh')
                    plt.title("DistribuciÃ³n de Consumos")
                    plt.xlabel("EnergÃ­a (Wh)")
                    plt.ylabel("Frecuencia")
                    plt.legend()
                    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
                    
                    # GrÃ¡fico 3: Consumo acumulado
                    plt.subplot(2, 2, 3)
                    plt.plot(df['Timestamp'], df['Energia Total (Wh)'], 'g-', linewidth=2)
                    plt.title("EnergÃ­a Total Acumulada")
                    plt.xlabel("Tiempo")
                    plt.ylabel("EnergÃ­a Total (Wh)")
                    plt.xticks(rotation=45)
                    plt.grid(True, linestyle='--', alpha=0.7)
                    
                    # GrÃ¡fico 4: Componentes temporales (STL)
                    if len(df) >= 20:
                        try:
                            plt.subplot(2, 2, 4)
                            stl = STL(df['Energia Evento (Wh)'].tail(50), period=5)
                            res = stl.fit()
                            fig = res.plot()
                            plt.title("DescomposiciÃ³n STL (Tendencia, Estacionalidad, Residual)")
                            plt.tight_layout()
                        except:
                            plt.subplot(2, 2, 4)
                            plt.text(0.5, 0.5, "STL no disponible\n(necesita mÃ¡s datos)", 
                                    ha='center', va='center')
                            plt.axis('off')
                    
                    plt.tight_layout()
                    plt.show()
                    
                    contador += 1

except KeyboardInterrupt:
    print("\nðŸ”´ Deteniendo monitor... Guardando datos finales...")
    df.to_csv("datos_energia_luz_final.csv", index=False)
    joblib.dump(config['modelos'][config['modelo_actual']], 'modelo_energia_final.pkl')
    arduino.close()
    print("âœ… Datos guardados correctamente. ConexiÃ³n cerrada.")

def generar_pdf(df):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    pdf.cell(200, 10, txt="REPORTE DE ENERGÃA", ln=True, align='C')
    
    pdf.ln(10)
    pdf.cell(200, 10, txt=f"Eventos registrados: {len(df)}", ln=True)
    pdf.cell(200, 10, txt=f"EnergÃ­a total: {df['Energia Total (Wh)'].iloc[-1]:.3f} Wh", ln=True)
    pdf.cell(200, 10, txt=f"Consumo promedio: {df['Energia Evento (Wh)'].mean():.3f} Wh", ln=True)
    
    pdf.ln(10)
    pdf.cell(200, 10, txt="Ãšltimos eventos:", ln=True)
    
    for i, fila in df.tail(5).iterrows():
        pdf.cell(200, 10, txt=f"{fila['Timestamp']} - {fila['Energia Evento (Wh)']:.3f} Wh", ln=True)
    
    pdf.output("reporte_energia.pdf")

generar_pdf(df)
```
